```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("ticks")
# sns.set(font_scale=1.5)
```

```python
data = pd.read_excel('./node reduction.xlsx')
data
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.scatterplot(data, x='% node reduction', y='% degradation', hue='city', legend="full")
    plt.legend(loc=(1.04, 0.4))
    plt.ylim(-103,5)
    plt.xlim(-5, 100)
    
    plt.savefig(f'figs/degradation_default_algo.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.scatterplot(data, x='tolerance', y='% node reduction', hue='city', legend="full")
    plt.legend(loc=(1.04, 0.4))
    # plt.ylim(-103,5)
    # plt.xlim(-5, 50)
    
    plt.savefig(f'figs/node reductions.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.scatterplot(data, x='tolerance', y='number of nodes coarsened graph', hue='city', legend="full")
    plt.legend(loc=(1.04, 0.4))
    # plt.ylim(-103,5)
    # plt.xlim(-5, 50)
    
    plt.savefig(f'figs/num nodes.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.lineplot(data, x='tolerance', y='number of nodes coarsened graph', hue='city', legend="full", marker='o', size=1)
    plt.legend(loc=(1.04, 0.4))
    # plt.ylim(-103,5)
    # plt.xlim(-5, 50)
    
    plt.savefig(f'figs/num nodes.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.lineplot(data, x='tolerance', y='% permutations', hue='city', marker='o', linewidth = 0.5, legend='full')
    # plt.legend(loc=(1.04, 0.4))
    # plt.ylim(0,0.1)
    # plt.xlim(-1, 50)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.savefig(f'figs/permutations.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.lineplot(data, x='tolerance', y='% permutations', hue='city', marker='o', linewidth = 0.5, legend='full')
    # plt.legend(loc=(1.04, 0.4))
    plt.ylim(0,0.1)
    plt.xlim(-1, 50)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.savefig(f'figs/permutations_zoom.png', bbox_inches='tight', dpi=300)
```
