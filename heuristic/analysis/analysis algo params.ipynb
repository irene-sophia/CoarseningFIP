```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("ticks")
# sns.set(font_scale=1.5)
```

```python
data = pd.read_excel('../results/node reduction.xlsx')
data
```

```python
data_default = data[data.weights == 'type']
# data_default = data_[data_['city'].isin(['Manhattan', 'Winterswijk', 'Utrecht'])]
data_default
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.scatterplot(data_default, x='% node reduction', y='% degradation', hue='city', style='pruning', size='iterations', legend="full")
    plt.legend(loc=(1.04, 0.4))
    plt.ylim(-103,5)
    # plt.xlim(-5, 50)
    
    plt.savefig(f'figs/degradation_default_algo.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full")
    sns.scatterplot(data_default, x='% node reduction', y='% degradation opt', hue='city', style='pruning', size='iterations', legend="full")
    plt.legend(loc=(1.04, 0.4))
    plt.ylim(-103,5)
    # plt.xlim(-5, 50)
    
    plt.savefig(f'figs/degradation_default_algo_opt.png', bbox_inches='tight', dpi=300)
```

```python
with sns.axes_style("whitegrid"):
    sns.scatterplot(data, x='% node reduction', y='% degradation', hue='weights', style='city', size='pruning', legend="full", sizes=[30,80])
    # sns.scatterplot(data, x='% node reduction', y='% degradation', hue='city', style='pruning', size='weights', legend="full")
    plt.legend(loc=(1.04, 0.4))
    plt.ylim(-103,5)
    plt.xlim(-5, 100)
    
    plt.savefig(f'figs/degradation_betweenness_and_default.png', bbox_inches='tight', dpi=300)
```
